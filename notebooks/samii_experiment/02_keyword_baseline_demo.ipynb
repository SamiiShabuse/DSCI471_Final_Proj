```python
import sys
from pathlib import Path

PROJECT_ROOT = Path("../..").resolve()
sys.path.insert(0, str(PROJECT_ROOT / "src"))

from baseline_keyword import build_keyword_index, search_keyword
```

```python
from PIL import Image
import matplotlib.pyplot as plt

def show_results(results):
    for _, row in results.iterrows():
        img_path = PROJECT_ROOT / row["image_path"]
        img = Image.open(img_path)

        plt.figure(figsize=(3, 3))
        plt.imshow(img)
        plt.axis("off")
        plt.title(row["productDisplayName"])
        plt.show()

        print("Text:", row["product_text"][:200], "...")
        print("Score:", round(row["similarity_score"], 4))
        print("-" * 50)
```

```python
PRODUCTS_CSV_PATH = PROJECT_ROOT / "data/processed/products.csv"

df, vectorizer, text_matrix = build_keyword_index(PRODUCTS_CSV_PATH)
print(f"Indexed {len(df):,} products")
```

```python
query = "blue casual men's shirt"
results = search_keyword(query, df, vectorizer, text_matrix, top_k=5)
show_results(results)
```

```python
results = search_keyword(
    "blue casual men's shirt",
    df,
    vectorizer,
    text_matrix,
    top_k=5
)

show_results(results)
```

```python
queries = [
    "flowy floral dress with cap sleeves",
    "black formal leather shoes",
    "red casual summer t shirt",
    "women party dress",
    "white running sneakers",
]

for query in queries:
    print("\nQuery:", query)
    display(search_keyword(query, df, vectorizer, text_matrix, top_k=5))
    show_results(search_keyword(query, df, vectorizer, text_matrix, top_k=5))
```
